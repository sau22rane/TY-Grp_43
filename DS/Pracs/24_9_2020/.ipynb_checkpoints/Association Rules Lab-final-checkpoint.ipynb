```python
import pandas as pd
import numpy as np
import math
```

```python
df = pd.read_csv("OnlineRetail.csv")
df.head()
```

```python
df.dtypes
```

```python
df['Description'] = df['Description'].str.strip()
df.dropna(axis=0, subset=['InvoiceNo'], inplace=True)
df['InvoiceNo'] = df['InvoiceNo'].astype('str')
df = df[~df['InvoiceNo'].str.contains('C')]
```

```python
# Set transactions
gp_invoiceno = df.groupby(['InvoiceNo','InvoiceDate'])
transactions = []
for name,group in gp_invoiceno:
    transactions.append(list(group['Description'].map(str)))
```

```python
transactions[0]
```

```python
from apyori import apriori
rules = apriori( transactions , min_support = 0.001, min_confidence = 0.8, max_length = 10)
```

```python
results = list(rules)
results
```

```python
print("There are {} RULES derived.".format(len(results)))
```

```python
for i in  range(10):
  item = results[i] 
  pair=item[0]
  #print(pair)
  items=[x for x in pair]
  #print (items)
  print(items[0]+'->'+items[1])
  # second index of the inner list
  print("Support: " + str(item[1]))
  # third index of the list located at 0th
  # of the third index of the inner list

  print("Confidence: " + str(item[2][0][2]))
  print("Lift: " + str(item[2][0][3]))
  print("==============",i+1,"=================")
```

```python
rules = apriori( transactions ,  min_confidence = 1)
```

```python
results = list(rules)
```

```python
print("There are {} RULES derived.".format(len(results)))
```
