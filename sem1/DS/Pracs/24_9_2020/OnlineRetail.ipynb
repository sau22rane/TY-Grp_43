```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("OnlineRetail.csv")
```

```python
df
```

```python
items = []
m = 0
lena = []
group = df.groupby('InvoiceNo')
for i,j in group:
    lena.append(len(j["Description"]))
    if(len(j["Description"])>400):
        continue
    temp = []
    for l in j["Description"]:
        temp.append(l)
    items.append(temp)
    if(len(j["Description"])>400):
        m+=1
print(len(items),m)
```

```python
print(items[-5:])
```

```python
#a = np.array(items)
df = pd.DataFrame(items)
len(df)
```

```python
df.to_csv("temp.csv")
```
