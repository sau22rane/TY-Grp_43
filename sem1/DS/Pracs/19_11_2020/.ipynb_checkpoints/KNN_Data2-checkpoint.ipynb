```python
import pandas as pd
import numpy as np
import math
```

```python
df_train = pd.read_csv("Train2.csv")
df_train.head(5)
```

```python
df_test = pd.read_csv("Test2.csv")
df_test.head(5)
```

```python
def most_frequent(List): 
    counter = 0
    num = List[0]
    
    for i in List:
        curr_frequency = List.count(i)
        if(curr_frequency> counter):
            counter = curr_frequency
            num = i
    return num
```

```python
def KNN(data, test, k = 3):
    _dist = {}
    for i in data:
        dist = 0
        for j in range(1,4):
            if(test[j] != i[j]):
                dist+=1
        _dist[dist] = i[-1]
    print(_dist)
    
    top = []
    count = 0
    for i in sorted (_dist.keys()):
        top.append(_dist[i])
        count+=1
        if(count == k):
            break
    print("closest:", top)
    print()
    return most_frequent(top)
```

```python
result = []
for i in range(len(df_test)):
    result.append(KNN(df_train.to_numpy(), df_test.to_numpy()[i]))
df_test.insert(len(df_test.columns),"Predicted Risk",result )
```
